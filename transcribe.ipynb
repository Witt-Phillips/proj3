# Part 1: Speech-to-Text Transcription

**Law and LLMs - Project 3**

Uses OpenAI's Whisper model to transcribe audio from the Depp v. Heard video.
Supports both: (1) YouTube URL (downloads + preprocesses per assignment), (2) local audio file.

```python
# Install: pip install -r requirements.txt
# Note: Works without ffmpeg on macOS (uses audioread + Core Audio for M4A)
from transcribe import transcribe_audio
from pathlib import Path
```

```python
# Configuration - use YouTube URL (assignment video) or local file
INPUT = "https://www.youtube.com/watch?v=IfN9F93xr_g"  # Assignment video; or "Depp v. Heard.wav"
OUTPUT_FILE = "Depp v. Heard_transcription.txt"
MODEL_NAME = "base.en"  # English-only; options: tiny.en, base.en, small.en, medium.en
```

```python
# Transcribe (YouTube: downloads + preprocesses first; file: loads directly)
transcription = transcribe_audio(
    audio_path_or_url=INPUT,
    output_path=OUTPUT_FILE,
    model_name=MODEL_NAME,
)
```

```python
# Display (already saved by transcribe_audio)
print("\n--- Transcription ---")
print(transcription)
```
